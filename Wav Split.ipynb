# WAV Segmentation

```python
from pydub import AudioSegment
```

```python
# import audio.wav file
Audio = AudioSegment.from_wav("ClaudioAudio.wav")

# generate several 
for i in range(0,len(Audio),333): # move by a quarter of second, thus the wav files will have some overlapping part, 
                                  # but more files can be generated by a shorter speech    
    newAudio = Audio[i:i+1000] # each index refers to a millisecond, hence [0:1000] is a second 
    newAudio.export('%d-2.wav'%(i/250), format="wav") #Exports to a wav file in the current path.
```
